```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
```

```python
model_id = "microsoft/Phi-3-mini-4k-instruct"
quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    #quantization_config=quant_config,
    device_map="auto",
    trust_remote_code=False
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
print(f"General Model Compute Dtype: {model.dtype}")
mem_gb = model.get_memory_footprint() / (1024**3)
print(f"\nQuantized Model Type: {type(model).__name__}")
print(f"Reduced Memory Footprint: {mem_gb:.2f} GB")
```

```python
prompt = "Instruct: Explain why the sky is blue in one sentence.\nOutput:"
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

outputs = model.generate(**inputs, max_new_tokens=50)
print("\n--- Model Output ---")
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
